```python
#巢狀if
#潤年測試：可被4整除，且除以100時餘數!=0或除以400時餘數=0
year=input("enter year: ")
year=int(year)
if (year%4==0):
    if (year%100!=0 or year%400==0):
        print("%d 是潤年"%year)
    else:
        print("%d 不是潤年"%year)
else:
    print("%d 不是潤年"%year)
```

```python
#exercise 1
import re
def is_number(num):
    pattern = re.compile(r'^[-+]?[-0-9]\d*\.\d*|[-+]?\.?[0-9]\d*$')
    #使用正規表示式(ch16)
    result = pattern.match(num)
    if result:
        return True
    else:
        return False

while(True):
    num=input("enter your number , q to quit:")
    if num=='q':
        break
    elif not (is_number(num)):
        print('error input')
    else:
        print(-int(num))
```

```python
#exercise 2
#大寫->小寫
#小寫->大寫
#數字->輸出
#其它->錯誤
import re
def is_number(num):
    pattern = re.compile(r'^[-+]?[-0-9]\d*\.\d*|[-+]?\.?[0-9]\d*$')
    #使用正規表示式(ch16)
    result = pattern.match(num)
    if result:
        return True
    else:
        return False
    
while(True):
    input_str=input("exit to quit.")
    if input_str=='exit':
        break
    else:
        if input_str.isupper():
            print(input_str.lower())
        elif input_str.islower():
            print(input_str.upper())
        elif is_number(input_str):
            print(input_str)
        else:
            print("error input")
```
