```python
class Banks():
    
    def __init__(self,uname):    #初始化方法
        self.__name = uname
        self.__balance = 0
        self.__title = "Taipei Bank"
        self.__rate = 30    # 預設美金與台幣換匯比例
        self.__service_charge = 0.01
        
    def save_money(self,money):      # 設計存款方式
        self.__balance += money       #  執行存款
        print("存款：" ,money," 完成")
        
    def withdraw_money(self,money):  #設計提款方式
        self.__balance -= money
        print("提款：",money," 完成")
        
    def get_balance(self):
        print(self.__name.title(),"目前餘額",self.__balance)
        
    def usa_to_taiwan(self,usa_d):      #美金況換台幣的方法
        self.result = self.__cal_rate(usa_d)  # 在類內部調用私有方法
        return self.result
    
    def __cal_rate(self,usa_d):    # 定義換匯是私有方法
        return int(usa_d * self.__rate * ( 1 - self.__service_charge))
        
hungbank = Banks("張三")
usdallor = 50
print(usdallor,"美金可以況換 ",hungbank.usa_to_taiwan(usdallor),"台幣")
```

13-2-5 衍生類引用基類的方法
衍生類引用基類的方法時需使用 super()，下列將使用另一類的類瞭解這個觀念。

例：這是一個衍生類調用基類方法的實例，筆者首先建立一個 Animals 類，然後建立這個類的衍生類 Dogs，Dogs 類在初始化中會使用 super() 凋用 Animals 類的初始化方法，可參考第 13 行，經過初始化處理後， mydog.name 將由 lily 變為 My pet lily。

```python
class Anmals():    # 基類
    def __init__(self,animal_name,animal_age):
        self.name = animal_name    # 記錄動物名稱
        self.age = animal_age      # 記錄動物年齡
        
    def run(self):
        print(self.name.title(),"正在走路")
        
class Dogs(Anmals):   # 衍生類
    def __init__(self,dog_name,dog_age):
        super().__init__("我的" + dog_name.title(),dog_age)
        
mycat = Anmals("露西",5)    # 建立 Animals 物件以及測試
print(mycat.name.title()," 今年 ",mycat.age," 歲")
mycat.run()

mydog = Dogs('麗麗',6)    # 建立 Dogs 物件以及測試
print(mydog.name.title()," 今年 ",mydog.age," 歲")
mydog.run()
```

12-3-6 三代同堂的類與取得基類的屬性 super()
在繼承觀念裡，我們也可以使用 Python 的 super() 方法取得基類的屬性，這對於設計三代同堂的類是很重要的。

下列是一個三代同堂的程式，在這個程式中有祖父（Grandfather）類，它的子類是父親（Father）類，父親類的子類是（Ivan）類。其實 Ivan 類要取得父親類的屬性很容易，可是要取得祖父類的屬性時就會碰上困難，解決方式是在 Father 類與 Ivan 類的 __init__( ) 方法中增加下列設定。

super( ) . __init__( ) # 將父類的屬性複製
這樣就可以使 Ivan 取得祖父類的屬性了。

例：這個程式會建立一個 Ivan 類的物件 ivan同，然後分別調用 Father 類和 Grandfather 類的方法列印資訊，接著分別取得 Father 類和 Grandfather 類的屬性。

```python
class Grandfather():
    '''定義祖父類'''
    def __init__(self):
        self.grandfathermoney = 10000
    def get_info1(self):
        print("Grandfather的信息")
        
class Father(Grandfather):
    '''定義父類'''
    def __init__(self):
        self.fathermoney = 8000
        super().__init__()
    def get_info2(self):
        print("Father的信息")
        
class Ivan(Father):
    '''定義子類'''
    def __init__(self):
        self.ivanmoney = 3000
        super().__init__()
    def get_info3(self):
        print("Ivan的信息")
    def get_money(self):
        print("\nIvan資產：",self.ivanmoney,
              "\n父親資產：",self.fathermoney,
              "\n祖父資產：",self.grandfathermoney)
        
ivan = Ivan()
ivan.get_info3()
ivan.get_info2()
ivan.get_info1()
ivan.get_money()
```

```python
class Father():
    def __init__(self):
        self.fathermoney = 10000
        
class Ira(Father):
    def __init__(self):
        self.iramoney = 800
        super().__init__()
        
class Ivan(Father):
    def __init__(self):
        self.ivanmoney = 3000
        super().__init__()
    def get_money(self):
        print("Ivan的資產：",self.ivanmoney,
            "\n父產資產  ：",self.fathermoney,
            "\nIra資產   ：",Ira().iramoney)
        
ivan = Ivan()
ivan.get_money()
```

12-4 多型（polymorphism）

```python
class Animals():
    def __init__(self,animal_name):
        self.name = animal_name      # 紀錄動物名稱
    def which(self):                 # 回傳動物名稱
        return "我的" + self.name.title()
    def action(self):                # 動物行為
        return "睡覺"

class Dogs(Animals):
    def __init__(self,dog_name):       # 紀錄動物名稱
        super().__init__(dog_name.title())
    def action(self):                  # 動物行為
        return "在街上奔跑"
    
class Monkeys():
    def __init__(self,monkey_name):     # 紀錄動物名稱
        self.name = "我的猴子" + monkey_name.title()
    def which(self):                     # 回傳動物名稱
        return self.name
    def action(self):                    # 動物行為
        return "在森林裡奔跑"
    
def doing(obj):      # 列出動物的行為
    print(obj.which()," 是在 ",obj.action())

my_cat = Animals("露西")
doing(my_cat)
my_dog = Dogs("吉米")
doing(my_dog)
my_monkey = Monkeys("泰勒")
doing(my_monkey)
```

12-7 特殊屬性
當設計或是看到別人設計的 Python 程式時，若是遇到 __xx__ 類的字串就要特別留意了，這些大多數是特殊屬性或方法，筆者將簡要說明幾個重要常見的屬性。

```python
def getMax(x,y):
    '''文檔字串實例
建議 x,y 是整數
這個函數將傳回羅大值'''
    if  int(x) > int(y):
        return x
    else:
        return y
print(getMax(2,3))
print(getMax.__doc__)  #列印文檔字串
```

12-8-3 __iter__() 方法
建立類的時候也可以將類定義成一個反覆運算物件，類似 list 或 tuple，供 for … in 迴圈內使用，這時類需設計 next() 方法，取得下一個值，直到達到結束條件，可以使用 rais StopIteration（第 15 章會解說，raise）終止繼續。

例：Fib 序列數的設計

```python
class Fib():
    def __init__(self,max):
        self.max = max
        
    def __iter__(self):
        self.a = 0
        self.b = 1
        return self
    
    def __next__(self):
        fib = self.a
        if fib > self.max:
            raise StopIteration
        self.a,self.b = self.b,self.a + self.b
        return fib
for i in Fib(100):
    print(i)
```
