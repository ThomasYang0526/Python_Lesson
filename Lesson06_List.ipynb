6-5-3 sorted() 排序
前一小節的 sort() 排序將造成清單元素順序永久更改，如果你不希望更改清單元素的順序，可以使用另一種排序 sorted()，使用這個排序可以獲得想要的排序結果，我們可以用新清單存儲新的排序清單，同時原先清單的順序不更改。它的使用如下：

new_list.sorted( name_list )

例：這個程式使用 car_sorted() 新清單存儲 car 清單的排序結果，同時使用 num_sorted 新清單存儲 num 清單的排序結果

```python
cars = ["honda","bmw","toyota","ford"]
print("目前 cars 清單的內容：",cars)
print("使用 sorted() 由小排到大")
cars_sorted = sorted(cars)
print("排序串列結果",cars_sorted)
print("原先串列 cars 內容：",cars)
num = [5,3,9,2]
print("目前 num 清單的內容：",num)
print("使用 sorted() 由小排到大")
num_sorted = sorted(num)
print("排序串列結果",num_sorted)
print("原先串列 num 內容：",num)
```

```python
cars = ["honda","bmw","toyota","ford"]
print("目前 cars 清單的內容：",cars)
print("使用 sorted(reverse=True) 由大排到小")
cars_sorted = sorted(cars,reverse=True)
print("排序串列結果",cars_sorted)
print("原先串列 cars 內容：",cars)
num = [5,3,9,2]
print("目前 num 清單的內容：",num)
print("使用 sorted(reverse()) 由大排到小")
num_sorted = sorted(num,reverse=True)
print("排序串列結果",num_sorted)
print("原先串列 num 內容：",num)
```
