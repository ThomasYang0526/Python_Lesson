```python
def mutifunction(x1,x2):
    '''加减乘除'''
    addresult = x1 + x2
    subresult = x1 - x2
    mulresult = x1 * x2
    divresult = x1 / x2
    return addresult,subresult,mulresult,divresult
x1 = x2 = 10
add,sub,mul,div = mutifunction(x1,x2)
print("加法结果：",add)
print("减法结果：",sub)
print("乘法结果：",mul)
print("除法结果：",div)
```

```python
def kitchen(unserved,served):
    ''' 将未服务的餐点转为已服务 '''
    print("厨房处理顾客所点的餐点")
    while unserved:
        current_meal = unserved.pop()
        #模拟出点餐过程
        print("菜单：",current_meal)
        #将已出餐点转入已经服务列表
        served.append(current_meal)
        
def show_unserved_meal(unserved):
    ''' 显示尚未服务的餐点 '''
    print("=== 下列是尚未服务的餐点 ===")
    if not unserved:
        print("*** 没有点餐 ***",'\n')
    for unserved_meal in unserved:
        print(unserved_meal)
        
def show_served_meal(served):
    ''' 显示已经服务的餐点 '''
    print("=== 下列是已经服务的餐点 ===")
    if not served:
        print("*** 没有餐点 ***",'\n')
    for served_meal in served:
        print(served_meal)
        
        
# 列出餐厅处理前的点餐内容
unserved = ['大麦克','劲辣鸡腿堡','麦克鸡块']
served = []

#列出餐厅处理后的点餐内容
show_unserved_meal(unserved)
show_served_meal(served)

# 餐厅服务过程
kitchen(unserved,served)
print('\n','=== 厨房处理结束 ===','\n')

#列出餐厅处理后的点餐内容
show_unserved_meal(unserved)
show_served_meal(served)
```

```python
def build_dict(name,age,**players):
    info = {}
    info["Name"] = name
    info["Age"] = age
    for key,value in players.items():
        info[key] =value
    return info

player_dict = build_dict("James",'32',
                        City = 'Cleveland',
                        State = 'Ohio')

print(player_dict)
```

11-8 匿名函数 lambda
匿名函数最大特色是可以有许多参数，但是只能有一个程序表达式，然后可以将执行结果返回。

lambda arg1 [ , arg2 , … ,argn ] : expression # agr1 是参数，可以有多个参数
例：这是单一参数的匿名函数应用，可以返回平方值。
匿名函数的最大优点是可以让程序变得更简洁，可参考下列实例.

```python
square = lambda x:x ** 2
print(square(10))
```

filter( function, iterable)
map ( function , iterable )

```python
def oddfn(x):
    return x if (x % 2 == 1) else None
mylist = [5,10,15,20,25,30]
filter_object = filter(oddfn,mylist)
print("奇数列表：",[item for item in filter_object])
```

```python
mylist = [5,10,15,20,25,30]
oddlist = list(filter(lambda x:(x % 2 == 1),mylist))
print("奇数列表：",oddlist)
```
