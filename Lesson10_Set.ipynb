在 Python 語言的交集符號是 “ & ” ，另外，也可以使用 intersection() 方法完成這個工作。

例：有數學與物理 2 個夏令營，這個程式會列出同時參加這 2 個夏令營的成員。

10-2-3 差集（difference）
有 A 和 B 兩個集合，如果想獲得屬於 A 集合元素，同時不屬於 B 集合則可以使用差集（A - B ）。如果想獲得屬於 B 集合元素，同時不屬於 A 集合則可以使用差集（B - A）。例如，有數學（可想成 A 集合）與物理（可想成 B 集合） 2 個夏令營，如果想瞭解參加數學夏令營但是沒有參加物理夏令營的成員，則可以使用此功能。

在 Python 語言的差集符號是 “ - ”，另外，也可以使用 difference() 方法完成這個工作。

例：有數學與物 2 個夏令營，這個程式會列出參加數學夏令營但是沒有參加物理夏令營所有成員，另處也會列出參加物理夏令營但是沒有參加數學夏令營的所有成員。

10-2-7 對稱差集（symmetric difference）
有 A 和 B 兩個集合，如果想獲得屬於 A 或 B 集合元素，但是排除同時屬於 A 和 B 的元素，則可以使用對稱差集。例如，有數學（可想成 A 集合）與物理（可想成 B 集合）2 個夏令營，如果想統計參加數學夏令營或是參加物理夏令營的成員，但是排除同時參加這 2 個夏令營的成員，則可以使用此功能。更簡單的解釋是只參加一個夏令營的成員。

在 Python 語言的對稱差集符號是 “ ^ ”，另外，也可使用 symmetric_difference() 方法完成這個工作。

例：有數學與物理 2 個夏令營，這個程式會列出參加數學夏令營或參加物理夏令營，但是排除同時參加 2 個夏令營的所有成員。

10-5 凍結集合 frozenset
set 是可變集合，frozenset 是不可變集合也可直譯為凍結集合，這是一個新的類別，只要設定元素後，這個凍結集合就不能再更改了。如果將元想成不可變清單，凍結集合就是不可變集合。

凍結集合的不可變特性優點是可以用它作字典的鍵，也可以作為其他集合的元素，凍結集合的建立方式是使用 frozenset() 函數，凍結集合建立完成後，不可使用 add() 或 remove() 更改凍結集合的內容，但是可以執行 intersection()、union()、difference()、symmetric_difference()、copy()、issubset()、issuperset()、isdisjoint() 等方法。

例：建立凍集合與氣操作。
