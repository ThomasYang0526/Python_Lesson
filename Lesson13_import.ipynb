```python
# ch13_14.py
import random    # 导入模块 random

n = 3
for i in range(n):
    print("1 - 100       ：",random.randint(1,100))

for i in range(n):
    print("500 - 1000    ：",random.randint(500,1000))

for i in range(n):
    print("2000 - 3000    ：",random.randint(2000,3000))
```

```python
# ch13_15.py
import random

min,max = 1,10
ans = random.randint(min,max)
while True:
    yourNum = int(input("请猜 1 - 10 之间数字："))
    if yourNum == ans:
        print("答对了")
        break
    elif yourNum < ans:
        print("请猜大一些")
    else:
        print("请猜小一些")
```
