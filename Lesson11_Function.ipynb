```python
def kitchen(unserved,served):
    ''' 將未服務的餐點轉為已服務 '''
    print("廚房處理顧客所點的餐點")
    while unserved:
        current_meal = unserved.pop()
        #模擬出點餐過程
        print("菜單：",current_meal)
        #將已出餐點轉入已經服務清單
        served.append(current_meal)
        
def show_unserved_meal(unserved):
    ''' 顯示尚未服務的餐點 '''
    print("=== 下列是尚未服務的餐點 ===")
    if not unserved:
        print("*** 沒有點餐 ***",'\n')
    for unserved_meal in unserved:
        print(unserved_meal)
        
def show_served_meal(served):
    ''' 顯示已經服務的餐點 '''
    print("=== 下列是已經服務的餐點 ===")
    if not served:
        print("*** 沒有餐點 ***",'\n')
    for served_meal in served:
        print(served_meal)
        
        
# 列出餐廳處理前的點餐內容
unserved = ['大麥克','勁辣雞腿堡','麥克雞塊']
served = []

#列出餐廳處理後的點餐內容
show_unserved_meal(unserved)
show_served_meal(served)

# 餐廳服務過程
kitchen(unserved,served)
print('\n','=== 廚房處理結束 ===','\n')

#列出餐廳處理後的點餐內容
show_unserved_meal(unserved)
show_served_meal(served)
```

11-8 匿名函數 lambda
匿名函數最大特色是可以有許多參數，但是只能有一個程式運算式，然後可以將執行結果返回。

lambda arg1 [ , arg2 , … ,argn ] : expression # agr1 是參數，可以有多個參數
例：這是單一參數的匿名函數應用，可以返回平方值。
匿名函數的最大優點是可以讓程式變得更簡潔，可參考下列實例.

```python
square = lambda x:x ** 2
print(square(10))
```

filter( function, iterable)

```python
def oddfn(x):
    return x if (x % 2 == 1) else None
mylist = [5,10,15,20,25,30]
filter_object = filter(oddfn,mylist)
print("奇數列表：",[item for item in filter_object])
```

```python
mylist = [5,10,15,20,25,30]
oddlist = list(filter(lambda x:(x % 2 == 1),mylist))
print("奇數列表：",oddlist)
```
