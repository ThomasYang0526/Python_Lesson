10-2-1 交集（intersection）
有 A 和 B 两个集合，如果想得相同的元素，则可以使用交集。例如，有数学（可想成 A 集合）与物理（可想成 B 集合）2 个夏令营，如果想统计有哪些人同时参加这 2 个夏令营，则可以使用此功能。

在 Python 语言的交集符号是 “ & ” ，另外，也可以使用 intersection() 方法完成这个工作。

例：有数学与物理 2 个夏令营，这个程序会列出同时参加这 2 个夏令营的成员。

10-2-2 并集（union）
有 A 和 B 两个集合，如果想获得所有的元素，则可以使用并集，例如，有数学（可想成 A 集合）与物理（可想成 B 集合）2 个夏令营，如果想统计参加这 2 个夏令营的全部成员，则可以使用 此功能

在 Python 语言的并集符号是 " | "，另外，也可以使用 union() 方法完成这个工作。

例：有数学与物理 2 个夏令营，这个程序会列出参加这 2 个夏令营的所有成员。

10-2-3 差集（difference）
有 A 和 B 两个集合，如果想获得属于 A 集合元素，同时不属于 B 集合则可以使用差集（A - B ）。如果想获得属于 B 集合元素，同时不属于 A 集合则可以使用差集（B - A）。例如，有数学（可想成 A 集合）与物理（可想成 B 集合） 2 个夏令营，如果想了解参加数学夏令营但是没有参加物理夏令营的成员，则可以使用此功能。

在 Python 语言的差集符号是 “ - ”，另外，也可以使用 difference() 方法完成这个工作。

例：有数学与物 2 个夏令营，这个程序会列出参加数学夏令营但是没有参加物理夏令营所有成员，另处也会列出参加物理夏令营但是没有参加数学夏令营的所有成员。
————————————————

原文链接：https://blog.csdn.net/ihuxh26556/article/details/109677665

```python
math = {'Kevin','Peter','Eric'}      # 设定参加数学夏令营成员
physics = {'Peter','Nelson','Tom'}   # 设定参加物理夏令营成员
math_only = math - physics
print("参加数学夏令营同时没有参加物理夏令营的成员：",math_only)
physics_only = physics - math
print("参加物理夏令营同时没有参加数学夏令营的成员：",physics_only)
```

```python
A = {1,2,3,4,5}
B = {3,4,5,6,7}
A_B = A.difference(B)
print("A - B 的差集是：",A_B)
B_A = B.difference(A)
print("B - A 的差集是：",B_A)
```

10-2-7 对称差集（symmetric difference）
有 A 和 B 两个集合，如果想获得属于 A 或 B 集合元素，但是排除同时属于 A 和 B 的元素，则可以使用对称差集。例如，有数学（可想成 A 集合）与物理（可想成 B 集合）2 个夏令营，如果想统计参加数学夏令营或是参加物理夏令营的成员，但是排除同时参加这 2 个夏令营的成员，则可以使用此功能。更简单的解释是只参加一个夏令营的成员。

在 Python 语言的对称差集符号是 “ ^ ”，另外，也可使用 symmetric_difference() 方法完成这个工作。

例：有数学与物理 2 个夏令营，这个程序会列出参加数学夏令营或参加物理夏令营，但是排除同时参加 2 个夏令营的所有成员。
————————————————

```python
math = {'Kevin','Peter','Eric'}      # 设定参加数学夏令营成员
physics = {'Peter','Nelson','Tom'}   # 设定参加物理夏令营成员
math_sydi_physics = math ^ physics
print("没有同时参加数学和物理夏令营的成员：",math_sydi_physics)
```

```python
A = {1,2,3,4,5}
B = {3,4,5,6,7}
A_sydi_B = A.symmetric_difference(B)
print("A和B的对称差集是：",A_sydi_B)
B_sydi_A = B.symmetric_difference(A)
print("B和A的对称差集是：",B_sydi_A)
```

10-5 冻结集合 frozenset
set 是可变集合，frozenset 是不可变集合也可直译为冻结集合，这是一个新的类别，只要设定元素后，这个冻结集合就不能再更改了。如果将元想成不可变列表，冻结集合就是不可变集合。

冻结集合的不可变特性优点是可以用它作字典的键，也可以作为其他集合的元素，冻结集合的建立方式是使用 frozenset() 函数，冻结集合建立完成后，不可使用 add() 或 remove() 更改冻结集合的内容，但是可以执行 intersection()、union()、difference()、symmetric_difference()、copy()、issubset()、issuperset()、isdisjoint() 等方法。

例：建立冻集合与气操作。
