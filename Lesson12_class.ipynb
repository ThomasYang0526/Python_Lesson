12-2-2 私有方法

```python
class Banks():
    
    def __init__(self,uname):    #初始化方法
        self.__name = uname
        self.__balance = 0
        self.__title = "Taipei Bank"
        self.__rate = 30    # 预设美金与台币换汇比例
        self.__service_charge = 0.01
        
    def save_money(self,money):      # 设计存款方式
        self.__balance += money       #  执行存款
        print("存款：" ,money," 完成")
        
    def withdraw_money(self,money):  #设计提款方式
        self.__balance -= money
        print("提款：",money," 完成")
        
    def get_balance(self):
        print(self.__name.title(),"目前余额",self.__balance)
        
    def usa_to_taiwan(self,usa_d):      #美金况换台币的方法
        self.result = self.__cal_rate(usa_d)  # 在类内部调用私有方法
        return self.result
    
    def __cal_rate(self,usa_d):    # 定义换汇是私有方法
        return int(usa_d * self.__rate * ( 1 - self.__service_charge))
        
hungbank = Banks("张三")
usdallor = 50
print(usdallor,"美金可以况换 ",hungbank.usa_to_taiwan(usdallor),"台币")
```

13-2-5 衍生类引用基类的方法
衍生类引用基类的方法时需使用 super()，下列将使用另一类的类了解这个观念。

例：这是一个衍生类调用基类方法的实例，笔者首先建立一个 Animals 类，然后建立这个类的衍生类 Dogs，Dogs 类在初始化中会使用 super() 凋用 Animals 类的初始化方法，可参考第 13 行，经过初始化处理后， mydog.name 将由 lily 变为 My pet lily。

```python
class Anmals():    # 基类
    def __init__(self,animal_name,animal_age):
        self.name = animal_name    # 记录动物名称
        self.age = animal_age      # 记录动物年龄
        
    def run(self):
        print(self.name.title(),"正在走路")
        
class Dogs(Anmals):   # 衍生类
    def __init__(self,dog_name,dog_age):
        super().__init__("我的" + dog_name.title(),dog_age)
        
mycat = Anmals("露西",5)    # 建立 Animals 对象以及测试
print(mycat.name.title()," 今年 ",mycat.age," 岁")
mycat.run()

mydog = Dogs('丽丽',6)    # 建立 Dogs 对象以及测试
print(mydog.name.title()," 今年 ",mydog.age," 岁")
mydog.run()
```

12-3-6 三代同堂的类与取得基类的属性 super()
在继承观念里，我们也可以使用 Python 的 super() 方法取得基类的属性，这对于设计三代同堂的类是很重要的。

下列是一个三代同堂的程序，在这个程序中有祖父（Grandfather）类，它的子类是父亲（Father）类，父亲类的子类是（Ivan）类。其实 Ivan 类要取得父亲类的属性很容易，可是要取得祖父类的属性时就会碰上困难，解决方式是在 Father 类与 Ivan 类的 __init__( ) 方法中增加下列设定。

super( ) . __init__( ) # 将父类的属性复制
这样就可以使 Ivan 取得祖父类的属性了。

例：这个程序会建立一个 Ivan 类的对象 ivan同，然后分别调用 Father 类和 Grandfather 类的方法打印信息，接着分别取得 Father 类和 Grandfather 类的属性。

```python
class Grandfather():
    '''定义祖父类'''
    def __init__(self):
        self.grandfathermoney = 10000
    def get_info1(self):
        print("Grandfather的信息")
        
class Father(Grandfather):
    '''定义父类'''
    def __init__(self):
        self.fathermoney = 8000
        super().__init__()
    def get_info2(self):
        print("Father的信息")
        
class Ivan(Father):
    '''定义子类'''
    def __init__(self):
        self.ivanmoney = 3000
        super().__init__()
    def get_info3(self):
        print("Ivan的信息")
    def get_money(self):
        print("\nIvan资产：",self.ivanmoney,
              "\n父亲资产：",self.fathermoney,
              "\n祖父资产：",self.grandfathermoney)
        
ivan = Ivan()
ivan.get_info3()
ivan.get_info2()
ivan.get_info1()
ivan.get_money()
```

```python
class Father():
    def __init__(self):
        self.fathermoney = 10000
        
class Ira(Father):
    def __init__(self):
        self.iramoney = 800
        super().__init__()
        
class Ivan(Father):
    def __init__(self):
        self.ivanmoney = 3000
        super().__init__()
    def get_money(self):
        print("Ivan的资产：",self.ivanmoney,
            "\n父产资产  ：",self.fathermoney,
            "\nIra资产   ：",Ira().iramoney)
        
ivan = Ivan()
ivan.get_money()
```

12-4 多型（polymorphism）

```python
class Animals():
    def __init__(self,animal_name):
        self.name = animal_name      # 纪录动物名称
    def which(self):                 # 回传动物名称
        return "我的" + self.name.title()
    def action(self):                # 动物行为
        return "睡觉"

class Dogs(Animals):
    def __init__(self,dog_name):       # 纪录动物名称
        super().__init__(dog_name.title())
    def action(self):                  # 动物行为
        return "在街上奔跑"
    
class Monkeys():
    def __init__(self,monkey_name):     # 纪录动物名称
        self.name = "我的猴子" + monkey_name.title()
    def which(self):                     # 回传动物名称
        return self.name
    def action(self):                    # 动物行为
        return "在森林里奔跑"
    
def doing(obj):      # 列出动物的行为
    print(obj.which()," 是在 ",obj.action())

my_cat = Animals("露西")
doing(my_cat)
my_dog = Dogs("吉米")
doing(my_dog)
my_monkey = Monkeys("泰勒")
doing(my_monkey)
```

12-7 特殊属性
当设计或是看到别人设计的 Python 程序时，若是遇到 __xx__ 类的字符串就要特别留意了，这些大多数是特殊属性或方法，笔者将简要说明几个重要常见的属性。

```python
def getMax(x,y):
    '''文档字符串实例
建议 x,y 是整数
这个函数将传回罗大值'''
    if  int(x) > int(y):
        return x
    else:
        return y
print(getMax(2,3))
print(getMax.__doc__)  #打印文档字符串
```

```python
class Myclass:
    '''文档字符串实例
Myclass 类别的应用'''
    def __init__(self,x):
        self.x = x
    def printMe(self):
        '''文档字符串实例
Myclass 类内 printMe 方法的应用'''
        print("Hi",self.x)
        
data = Myclass(100)
data.printMe()
print(data.__doc__)
print(data.printMe.__doc__)
```

12-8-3 __iter__() 方法
建立类的时候也可以将类定义成一个迭代对象，类似 list 或 tuple，供 for … in 循环内使用，这时类需设计 next() 方法，取得下一个值，直到达到结束条件，可以使用 rais StopIteration（第 15 章会解说，raise）终止继续。

例：Fib 序列数的设计

```python
class Fib():
    def __init__(self,max):
        self.max = max
        
    def __iter__(self):
        self.a = 0
        self.b = 1
        return self
    
    def __next__(self):
        fib = self.a
        if fib > self.max:
            raise StopIteration
        self.a,self.b = self.b,self.a + self.b
        return fib
for i in Fib(100):
    print(i)
```
