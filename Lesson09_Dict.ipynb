```python
#將清單轉成字典
seq1 = ['name','city']
list_dict1 = dict.fromkeys(seq1)
print("字典1：",list_dict1)
list_dict2 = dict.fromkeys(seq1,'Chicago')
print("字典2：",list_dict2)

#將元組轉成字典
seq2 = ('name','city')
tup_dict1 = dict.fromkeys(seq2)
print("字典3：",tup_dict1)
tup_dict2 = dict.fromkeys(seq2,"New York")
print("字典4",tup_dict2)
```

```python
fruits = {'Apple':20,'Orange':25}
ret_value1 = fruits.get('Orange')  # 找到鍵就返回此鍵的值
print("Value = ",ret_value1)
ret_value2 = fruits.get("Grang")   # 找不到鍵就返回 None
print("Value = ",ret_value2)
ret_value3 = fruits.get("Grape",10) # 找不到鍵就返回第2個參數，即預設值
print("Value = ",ret_value3)
```

```python
person = {'name':'John'}
print("原先字典內容：",person)

#'age'鍵不存在
age = person.setdefault('age')
print("增加 age 鍵",person)
print("age = ",age)

#'sex'鍵不存在
sex = person.setdefault('sex','Male')
print("增加 sex 鍵",person)
print("sex = ",sex)
```
